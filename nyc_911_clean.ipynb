```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("local[*]") \
    .appName("NYC911_Local") \
    .getOrCreate()

df = spark.read \
    .option("header", "true") \
    .option("sep", ",") \
    .option("quote", '"') \
    .option("escape", '"') \
    .option("multiLine", "true") \
    .csv("/Users/sayanibrahmachari/Downloads/nyc_911.csv")

print("Number of columns:", len(df.columns))
print(df.columns)
df.show(5, truncate=False)
```

```python
df.printSchema()
```

```python
df.show(5)
```

```python
cols = [
    "CAD_EVNT_ID",
    "INCIDENT_DATE",
    "INCIDENT_TIME",
    "BORO_NM",
    "TYP_DESC",
    "Latitude",
    "Longitude"
]

df.select(*cols).show(10, truncate=False)
```

```python
df = df.toDF(*[c.strip().replace('"', '') for c in df.columns])
print(df.columns)
```

Check nulls

```python
from pyspark.sql.functions import col, sum as spark_sum, trim

null_counts = df.select([
    spark_sum(
        (col(c).isNull() | (trim(col(c)) == "")).cast("int")
    ).alias(c)
    for c in df.columns
])

null_counts.show(vertical=True, truncate=False)
```

```python
df_clean = df.dropna(subset=["BORO_NM", "TYP_DESC", "INCIDENT_DATE"])

before_rows = df.count()
after_rows = df_clean.count()

print(f"Before cleaning: {before_rows:,}")
print(f"After cleaning: {after_rows:,}")
print(f"Rows dropped: {before_rows - after_rows:,}")
```

```python
from pyspark.sql.functions import unix_timestamp, round as spark_round, col

# Keep rows where dispatch and arrival timestamps exist
df_emergency = df_clean.filter(
    col("ARRIVD_TS").isNotNull() &
    col("DISP_TS").isNotNull() &
    (col("ARRIVD_TS") != "") &
    (col("DISP_TS") != "")
)

# Calculate response time in minutes
df_emergency = df_emergency.withColumn(
    "response_time_mins",
    spark_round(
        (
            unix_timestamp(col("ARRIVD_TS"), "MM/dd/yyyy hh:mm:ss a") -
            unix_timestamp(col("DISP_TS"), "MM/dd/yyyy hh:mm:ss a")
        ) / 60,
        2
    )
)

# Remove impossible / bad values
df_emergency = df_emergency.filter(
    (col("response_time_mins") > 0) &
    (col("response_time_mins") < 120)
)

print(f"Emergency rows: {df_emergency.count():,}")
```

```python
from pyspark.sql.functions import when, hour, dayofweek, to_timestamp, col

df_emergency = df_emergency \
    .withColumn("hour_of_day",
        hour(to_timestamp(col("INCIDENT_TIME"), "HH:mm:ss"))) \
    .withColumn("day_of_week",
        dayofweek(to_timestamp(col("INCIDENT_DATE"), "MM/dd/yyyy"))) \
    .withColumn("is_night",
        when((col("hour_of_day") >= 22) | (col("hour_of_day") <= 6), 1)
        .otherwise(0)) \
    .withColumn("is_weekend",
        when(col("day_of_week").isin([1, 7]), 1)
        .otherwise(0)) \
    .withColumn("slow_response",
        when(col("response_time_mins") > 10, 1)
        .otherwise(0))
```

```python
df_emergency.select(
    "BORO_NM",
    "TYP_DESC",
    "hour_of_day",
    "day_of_week",
    "is_night",
    "is_weekend",
    "response_time_mins",
    "slow_response"
).show(10, truncate=False)
```

```python
df_emergency.createOrReplaceTempView("incidents")
```

```python
spark.sql("""
    SELECT
        BORO_NM,
        COUNT(*) as total_incidents,
        ROUND(AVG(response_time_mins), 2) as avg_response_mins,
        SUM(slow_response) as slow_count
    FROM incidents
    WHERE BORO_NM IS NOT NULL
    GROUP BY BORO_NM
    ORDER BY avg_response_mins DESC
""").show()
```

```python
spark.sql("""
    SELECT
        hour_of_day,
        COUNT(*) as total,
        ROUND(AVG(response_time_mins), 2) as avg_response,
        ROUND(SUM(slow_response) * 100.0 / COUNT(*), 2) as slow_pct
    FROM incidents
    GROUP BY hour_of_day
    ORDER BY hour_of_day
""").show(24)
```

```python
spark.sql("""
    SELECT
        slow_response,
        COUNT(*) as count
    FROM incidents
    GROUP BY slow_response
""").show()
```

```python
df_emergency.write \
    .mode("overwrite") \
    .parquet("/Users/sayanibrahmachari/Downloads/nyc_911_clean.parquet")

print("Saved clean parquet locally")
```
